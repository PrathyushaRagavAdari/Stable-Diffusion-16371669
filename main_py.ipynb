```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image, ImageDraw, ImageFont
import os

# --- 1. Setup Pipeline ---
print("Loading Stable Diffusion Pipeline...")
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda") # Ensure you are using a GPU

# --- 2. Data-to-Prompt Mapping ---
data_inputs = [
    {"animal": "Dog", "breed": "Golden Retriever", "condition": "Recovering from leg surgery, wearing a cone", "environment": "Bright veterinary clinic examination room"},
    {"animal": "Cat", "breed": "Maine Coon", "condition": "Healthy, receiving a routine checkup", "environment": "Cozy home living room"}
]

# --- 3. Control Strategy: Prompt Templates & Negative Prompts ---
def generate_naive_prompt(data):
    return f"A picture of a {data['breed']} {data['animal']} {data['condition']} in a {data['environment']}."

def generate_structured_prompt(data):
    # Uses specific keywords for higher quality and better adherence
    return (f"Educational veterinary illustration, high realism, 8k resolution, detailed photography: "
            f"A {data['breed']} {data['animal']}. Medical context: {data['condition']}. "
            f"Setting: {data['environment']}. "
            f"Soft lighting, compassionate tone, highly detailed texture, clinical accuracy.")

negative_prompt = "distorted, deformed, bad anatomy, unrealistic, blurry, cartoon, graphic injuries, blood, distressed animal, watermark, text"

# --- 4. Ethics Requirement: Watermarking ---
def add_ethics_watermark(image):
    draw = ImageDraw.Draw(image)
    text = "AI-Generated Illustration for Educational Purposes"
    # Fallback to default font if specific font isn't available
    draw.text((10, image.height - 20), text, fill=(255, 255, 255))
    return image

# --- 5. Generation and Evaluation Loop ---
os.makedirs("output_images", exist_ok=True)

for i, data in enumerate(data_inputs):
    print(f"\nProcessing Scenario {i+1}: {data['breed']} {data['animal']}")

    naive_prompt = generate_naive_prompt(data)
    structured_prompt = generate_structured_prompt(data)

    print(f"Naive Prompt: {naive_prompt}")
    print(f"Structured Prompt: {structured_prompt}")

    # Generate Naive Image (Baseline)
    print("Generating Baseline...")
    baseline_img = pipe(naive_prompt, num_inference_steps=20).images[0]
    baseline_img = add_ethics_watermark(baseline_img)
    baseline_img.save(f"output_images/scenario_{i+1}_baseline.png")

    # Generate Controlled Image (Improved)
    print("Generating Controlled Output...")
    improved_img = pipe(structured_prompt, negative_prompt=negative_prompt, num_inference_steps=30, guidance_scale=7.5).images[0]
    improved_img = add_ethics_watermark(improved_img)
    improved_img.save(f"output_images/scenario_{i+1}_improved.png")

print("\nGeneration complete. Check the 'output_images' folder.")
```
